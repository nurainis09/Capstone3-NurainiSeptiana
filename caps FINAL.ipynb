## Pemahaman Permasalahan Bisnis

### **Konteks**  
Jenis produk keuangan yang digunakan oleh masyarakat semakin beragam. Salah satu produk keuangan yang sudah dikenal luas oleh masyarakat adalah deposito berjangka. Mekanisme deposito berjangka adalah nasabah menyetorkan sejumlah uang ke bank atau lembaga keuangan, dan uang tersebut hanya dapat ditarik setelah jangka waktu tertentu. Sebagai kompensasi, nasabah akan mendapatkan bunga tetap sesuai dengan jumlah uang yang disetorkan.

Namun demikian, sebagai entitas bisnis dengan produk keuangan dan nasabah yang berbeda, bank tetap harus bersaing agar tidak kehilangan nasabah. Salah satu cara untuk mendapatkan nasabah baru adalah dengan melakukan kampanye pemasaran.

Target :

- 1 (True/Yes): Nasabah membuka deposito berjangka.
- 0 (False/No): Nasabah tidak membuka deposito berjangka.

### **Pernyataan Masalah :**

Meskipun produk deposito berjangka sudah dikenal luas, bank tetap harus bersaing untuk menarik nasabah baru dan mempertahankan nasabah yang sudah ada. Salah satu cara untuk mencapai hal ini adalah dengan melakukan kampanye pemasaran yang efektif. Namun, tidak semua nasabah akan tertarik untuk membuka deposito baru, dan tidak semua respons terhadap kampanye pemasaran berujung pada pembukaan produk.

### **Tujuan :**

Tujuan dari permasalahan ini adalah untuk membangun model prediksi yang dapat memprediksi kemungkinan nasabah atau calon nasabah akan membuka deposito berjangka setelah terpapar oleh kampanye pemasaran. Dengan menggunakan data yang tersedia, seperti karakteristik demografis, riwayat transaksi, interaksi dengan kampanye pemasaran, dan faktor keuangan lainnya, model ini bertujuan untuk membantu bank:
Mengidentifikasi nasabah yang memiliki kemungkinan tinggi untuk membuka deposito berjangka.
Menyusun strategi pemasaran yang lebih terarah dan efisien, yang akan meningkatkan konversi dari calon nasabah menjadi nasabah deposito.
Model ini dapat digunakan untuk menargetkan nasabah yang lebih tepat melalui kampanye pemasaran yang dipersonalisasi, meningkatkan pendapatan bank melalui peningkatan jumlah deposito yang dibuka, dan menurunkan biaya pemasaran dengan mengurangi upaya yang tidak efektif.

### **Evaluasi Metrik:**

False Positives (FP): Jumlah nasabah yang tidak membuka deposito, tetapi diprediksi akan membuka deposito.
Konsekuensi:
- Pemborosan biaya pemasaran.
- Pengiriman promosi yang tidak relevan.
- Penurunan kepercayaan nasabah.

False Negatives (FN): Jumlah nasabah yang membuka deposito, tetapi diprediksi tidak akan membuka deposito.
Konsekuensi:
- Kehilangan peluang pemasaran.
- Pendapatan yang hilang.
- Pengalaman nasabah yang buruk.
- Efisiensi kampanye yang rendah.

Sesuai dengan fakta yang terjadi [(sumber)](https://www.emarketer.com/content/some-bank-marketers-budgets-increased-2023-returns), bahwa peningkatan biaya pemasaran berbanding lurus dengan yang dihasilkan, maka untuk kampanye pemasaran bank yang mengutamakan mengurangi false negatives, metrik yang paling sesuai adalah Recall. Dengan recall tinggi, kita akan mengidentifikasi lebih banyak nasabah yang berpotensi membuka deposito, meskipun beberapa orang yang tidak tertarik tetap diprediksi tertarik.
Keuntungan bisnis utama: lebih banyak nasabah yang membuka deposito, yang berujung pada peningkatan jumlah dana yang ditempatkan di bank serta keuntungan pada tahap selanjutnya melalui Net Interest Margin.

Dengan perhitungan sebagai berikut:
- Bank atau lembaga keuangan menghabiskan biaya marketing minimum $ 20 per nasabah [sumber](https://www.bolddesk.com/blogs/customer-acquisition-cost). Indeks Biaya Hidup di Indonesia bisa sekitar 30% hingga 40% dari biaya hidup di AS, maka perhitungannya akan menjadi 40% dari $20 = $8 di Indonesia atau sebesar Rp. 132,000.- dengan rate Rp. 16,500,- per USD.
- Penempatan deposito berjangka melalui bank-bank yang ada di Indonesia rata-rata besaran minimumnya adalah Rp. 10,000,000,- [sumber](https://www.bankmandiri.co.id/deposito-rupiah-mandiri)
- Suku bunga pinjaman sesuai data OJK per Agustus 2024 [sumber](https://www.ojk.go.id/en/kanal/perbankan/data-dan-statistik/statistik-perbankan-indonesia/Pages/Indonesia-Banking-Statistic---August-2024.aspx) adalah 5.85% dan suku bunga kredit sebesar 10.10%, maka selisihnya adalah 4.25%.
- Apabila dari data didapatkan sebanyak 3,732 yang membuka deposito berjangka dari total nasabah sebanyak 7,813. maka total penempatan deposito berjangka adalah 37,320,000,000,-, dimana dikonversi menjadi Net Interest Margin sebesar 4.25% yaitu Rp. 1,586,100,000,- dan NIM setelah dibagi per nasabah adalah Rp. 425,000,-, maka dalam hal ini NIM per nasabah lebih besar dari pada biaya pemasaran yang dikorbankan. 


$$\ \text{Rasio} = \frac{\text{Keuntungan}}{\text{Biaya pemasaran}} \$$
$$\ \text{Rasio} = \frac{425,000}{132,000} =\text{3.22} \$$

Rasio ini menunjukkan bahwa keuntungan yang diperoleh lebih dari **3 kali lipat** dari biaya pemasaran yang dikeluarkan.

```python
categorical_columns = ['job', 'housing', 'loan', 'contact', 'month', 'poutcome']

plt.figure(figsize=(12, 10))

for i, col in enumerate(categorical_columns, 1):
    plt.subplot(2, 3, i)
    sns.countplot(data=df, x=col, hue='deposit')
    plt.title(f'Distribusi {col} berdasarkan Deposit')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# 2. Tabel Kontingensi untuk menghitung frekuensi dari variabel kategorikal berdasarkan 'deposit'
for col in categorical_columns:
    print(f"\nTabel Kontingensi untuk {col}:")
    contingency_table = pd.crosstab(df[col], df['deposit'])
    print(contingency_table)

# 3. Uji Chi-Square untuk menguji hubungan antara variabel kategorikal dan target 'deposit'
for col in categorical_columns:
    print(f"\nUji Chi-Square untuk kolom {col}:")
    contingency_table = pd.crosstab(df[col], df['deposit'])
    chi2, p_value, dof, expected = chi2_contingency(contingency_table)
    print(f"p-value: {p_value:.4f}")
    if p_value < 0.05:
        print("Kesimpulan: Ada hubungan yang signifikan antara", col, "dan target 'deposit'.")
    else:
        print("Kesimpulan: Tidak ada hubungan yang signifikan antara", col, "dan target 'deposit'.")
```

```python
# Fungsi untuk menghitung Cramr's V
def cramers_v(confusion_matrix):
    chi2_stat, p_val, dof, expected = chi2_contingency(confusion_matrix)
    n = confusion_matrix.sum()
    phi2 = chi2_stat / n
    r, k = confusion_matrix.shape
    return np.sqrt(phi2 / min(k - 1, r - 1))

contingency_table = pd.crosstab(df['job'], df['deposit'])
print("Cramr's V untuk job dan deposit:", cramers_v(contingency_table))

contingency_table = pd.crosstab(df['housing'], df['deposit'])
print("Cramr's V untuk housing dan deposit:", cramers_v(contingency_table))

contingency_table = pd.crosstab(df['loan'], df['deposit'])
print("Cramr's V untuk loan dan deposit:", cramers_v(contingency_table))

contingency_table = pd.crosstab(df['contact'], df['deposit'])
print("Cramr's V untuk contact dan deposit:", cramers_v(contingency_table))

contingency_table = pd.crosstab(df['month'], df['deposit'])
print("Cramr's V untuk month dan deposit:", cramers_v(contingency_table))

contingency_table = pd.crosstab(df['poutcome'], df['deposit'])
print("Cramr's V untuk poutcome dan deposit:", cramers_v(contingency_table))
```

## Pencocokan Model

### Penerapan Skala, Transformasi dan Model

```python
# Kolom kategorikal dan numerikal
categorical_cols = ['job', 'housing', 'loan', 'contact', 'month', 'poutcome']
numerical_cols = ['age', 'balance', 'pdays', 'campaign']

# Membuat transformasi kolom numerikal dan kategorikal secara terpisah
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), categorical_cols),
        ('num', RobustScaler(), numerical_cols)
    ])
```

```python
# Definisikan model-model yang akan digunakan
logreg = LogisticRegression(random_state=1515)
knn = KNeighborsClassifier()
dt = DecisionTreeClassifier(random_state=1515)
rf = RandomForestClassifier(random_state=1515)
ada = AdaBoostClassifier(random_state=1515)
gb = GradientBoostingClassifier(random_state=1515)
xgb = XGBClassifier(random_state=1515)
lgb = LGBMClassifier(random_state=1515, verbose=-1)
cat = CatBoostClassifier(verbose=False, random_state=1515)

models = [logreg, knn, dt, rf, ada, gb, xgb, lgb, cat]
```

### Validasi Silang

Cross validation atau validasi silang, dilakukan pada training set, untuk mengevaluasi model dan memahami seberapa baik model bekerja pada data yang tidak dilihat sebelumnya

```python
list_mean = []
list_std = []
model_name = []

for model in models:
    # Definisikan pipeline
    model_pipeline = Pipeline([
        ('preprocessing', preprocessor),
        ('model', model)
    ])

    # Cross-validation untuk mendapatkan skor recall
    model_cv = cross_val_score(
        model_pipeline,
        X_train,  # Pastikan X_train sudah benar
        y_train,  # y_train sudah di-label encoding
        cv=5,
        scoring='recall',  # Ganti scoring sesuai kebutuhan (misalnya recall, f1, accuracy)
        n_jobs=-1
    )

    # Menyimpan rata-rata dan standar deviasi hasil cross-validation
    list_mean.append(model_cv.mean())
    list_std.append(model_cv.std())
    model_name.append(model.__class__.__name__)

# Membuat DataFrame untuk menampilkan hasil
cv = pd.DataFrame({
    'model': model_name,
    'mean_recall': list_mean,
    'std_recall': list_std,
}).sort_values('mean_recall', ascending=False)  # Pastikan sort by 'mean_recall'

# Menampilkan hasil cross-validation
cv
```

### Hyperparameter Tuning

Hyperparameter tuning dilakukan menggunakan validation set (atau dengan cross validation yang dilakukan pada training set) untuk memilih model dengan hyperparameter terbaik.

#### Hyperparameter Tuning on Test Set

```python
scores = []
model_name = []

for model in models:
    model_pipeline = Pipeline([
        ('preprocessing', preprocessor),
        ('model', model)
    ])

    model_pipeline.fit(X_train,y_train)
    y_pred = model_pipeline.predict(X_test)

    score = recall_score(y_test,y_pred)
    scores.append(score)
    model_name.append(model.__class__.__name__)

val = pd.DataFrame({
    'model' : model_name,
    'Recall score (validation set)' : list_mean,
    'Recall score (test set)' : scores
}).sort_values('Recall score (test set)', ascending=False)

val
```

#### Hyperparameter Tuning: Random Forest

```python
model = RandomForestClassifier(random_state=1515)

# Parameter grid untuk RandomForestClassifier
rf_params = {
    'model__n_estimators': [50, 100, 200],
    'model__max_depth': [None, 10, 20, 30],
    'model__min_samples_split': [2, 5, 10],
    'model__min_samples_leaf': [1, 2, 4],
    'model__bootstrap': [True, False]
}

# Definisikan Pipeline
model_pipeline = ImbPipeline([
    ('preprocessing', preprocessor),  # Preprocessor yang sudah didefinisikan
    ('model', model)
])

# Lakukan GridSearchCV untuk RandomForestClassifier
grid_search = GridSearchCV(
    model_pipeline,
    rf_params,
    cv=5,  # Cross-validation dengan 5 fold
    scoring='recall',  # Gunakan metrik Recall
    n_jobs=-1,
    verbose=1
)

# Latih GridSearchCV
grid_search.fit(X_train, y_train)

# Ambil model terbaik dari GridSearchCV
rf_tuned = grid_search.best_estimator_

# Evaluasi model terbaik pada test set
y_pred_rf = rf_tuned.predict(X_test)

# Hitung skor Recall
rf_tuned_recall_score = recall_score(y_test, y_pred_rf)
rf_train_recall_score = cross_val_score(rf_tuned, X_train, y_train, cv=5, scoring='recall').mean()

# Menampilkan hasil evaluasi
print("Best Hyperparameters:", grid_search.best_params_)
print("\nRecall Score on Test Set:", rf_tuned_recall_score)
print("\nTrain Recall Score (CV):", rf_train_recall_score)
```

#### Hyperparameter Tuning : Decision Tree

```python
model = DecisionTreeClassifier(random_state=1515)

# Parameter grid untuk DecisionTreeClassifier
dt_params = {
    'model__criterion': ['gini', 'entropy'],  # Fungsi split
    'model__max_depth': [None, 10, 20, 30],  # Kedalaman maksimum
    'model__min_samples_split': [2, 5, 10],  # Minimum sampel untuk split
    'model__min_samples_leaf': [1, 2, 4],    # Minimum sampel per leaf
    'model__max_features': [None, 'sqrt', 'log2']  # Fitur maksimum saat split
}

# Definisikan Pipeline
model_pipeline = ImbPipeline([
    ('preprocessing', preprocessor),  # Preprocessor yang sudah didefinisikan
    ('model', model)
])

# Lakukan GridSearchCV untuk DecisionTreeClassifier
grid_search = GridSearchCV(
    model_pipeline,
    dt_params,
    cv=5,  # Cross-validation dengan 5 fold
    scoring='recall',  # Gunakan metrik Recall
    n_jobs=-1,
    verbose=1
)

# Latih GridSearchCV
grid_search.fit(X_train, y_train)

# Ambil model terbaik dari GridSearchCV
dt_tuned = grid_search.best_estimator_

# Evaluasi model terbaik pada test set
y_pred_dt = dt_tuned.predict(X_test)

# Hitung skor Recall
dt_tuned_recall_score = recall_score(y_test, y_pred_dt)
dt_train_recall_score = cross_val_score(dt_tuned, X_train, y_train, cv=5, scoring='recall').mean()

# Menampilkan hasil evaluasi
print("Best Hyperparameters:", grid_search.best_params_)
print("\nRecall Score on Test Set:", dt_tuned_recall_score)
print("\nTrain Recall Score (CV):", dt_train_recall_score)
```

#### Hyperparameter Tuning : LGBM Classifier

```python
model = LGBMClassifier(random_state=1515)

# Parameter grid untuk LGBMClassifier
lgbm_params = {
    'model__n_estimators': [50, 100, 200],
    'model__learning_rate': [0.01, 0.1, 0.2],
    'model__max_depth': [3, 4, 5],
    'model__num_leaves': [31, 63, 127],
    'model__min_child_samples': [20, 40, 60]
}

# Definisikan Pipeline
model_pipeline = ImbPipeline([
    ('preprocessing', preprocessor),  # Preprocessor yang sudah didefinisikan
    ('model', model)
])

# Lakukan GridSearchCV untuk LGBMClassifier
lgbm_grid_search = GridSearchCV(
    model_pipeline,
    lgbm_params,
    cv=5,  # Cross-validation dengan 5 fold
    scoring='recall',  # Gunakan metrik Recall
    n_jobs=-1,
    verbose=1
)

# Latih GridSearchCV
lgbm_grid_search.fit(X_train, y_train)

# Ambil model terbaik dari GridSearchCV
lgbm_tuned = lgbm_grid_search.best_estimator_

# Evaluasi model terbaik pada test set
y_pred_lgbm = lgbm_tuned.predict(X_test)

# Hitung skor Recall
lgbm_tuned_recall_score = recall_score(y_test, y_pred_lgbm)
lgbm_train_recall_score = cross_val_score(lgbm_tuned, X_train, y_train, cv=5, scoring='recall').mean()

# Menampilkan hasil evaluasi
print("Best Hyperparameters for LGBMClassifier:", lgbm_grid_search.best_params_)
print("\nRecall Score on Test Set for LGBMClassifier:", lgbm_tuned_recall_score)
print("\nTrain Recall Score (CV):", lgbm_train_recall_score)
```

Algoritma LIGHTGBM singkatan dari Light Gradient Boosting Machine, adalah sebuah algoritma machine learning yang menggunakan teknik gradient boosting untuk melakukan prediksi. Algoritma ini dikembangkan oleh Microsoft dan menjadi salah satu algoritma yang populer dalam dunia data  science. algoritma ini memiliki keunggulan dalam kecepatan dan efisiensi yang tinggi, serta mampu mengatasi masalah yang berkaitan dengan data yang besar dan fitur yang banyak.

Terdapat beberapa kelebihan yang dimiliki oleh algoritma Light Gradient Boosting Machine, antara lain:

1. Kecepatan yang Tinggi: LIGHTGBM menggunakan pendekatan yang berbeda dalam membangun pohon keputusan, sehingga mampu bekerja dengan cepat dan efisien. Algoritma ini dapat menghandle data yang besar dengan waktu komputasi yang lebih singkat dibandingkan dengan algoritma gradient boosting lainnya.

2. Efisiensi Penggunaan Memori: LIGHTGBM menggunakan teknik kompresi histogram yang memungkinkan penggunaan memori yang lebih efisien. Hal ini membuat algoritma ini cocok untuk pengolahan data dengan ukuran yang besar.

3. Mendukung Fitur Kategorikal: LIGHTGBM secara native mendukung fitur kategorikal tanpa perlu melakukan proses encoding terlebih dahulu. Hal ini memudahkan pengguna dalam mengolah data yang memiliki fitur kategorikal.

4. Skalabilitas: Algoritma LIGHTGBM dapat bekerja dengan baik pada dataset yang memiliki jumlah fitur yang banyak. Algoritma ini mampu mengatasi masalah yang berkaitan dengan dimensi yang tinggi.

Cara kerja LightGBM:

1. **Pembagian Data Menjadi Binning**: membagi setiap fitur numerik menjadi beberapa bin diskrit. Binning ini membantu dalam mengurangi kompleksitas perhitungan dan mempercepat proses pembelajaran.
2. **Membuat Histogram**: LightGBM membangun histogram untuk setiap fitur. Histogram ini merepresentasikan distribusi data dalam setiap bin.
3. **Fitur Terbaik untuk Splitting**: Gradient-based One-Side Sampling (GOSS) untuk mempercepat proses pemilihan fitur. GOSS memilih sebagian data yang memiliki gradient besar dan data acak yang memiliki gradient kecil.
4. **Pembentukan Pohon Keputusan**: LightGBM membangun banyak pohon keputusan secara paralel. Setiap pohon dibangun dengan menggunakan pendekatan penambahan leaf ke leaf (leaf-wise).Pendekatan leaf-wise memungkinkan LightGBM untuk menggabungkan banyak leaf yang dapat memperluas ruang solusi dan meningkatkan kemampuan generalisasi.
5. **Pemangkasan Pohon**: LightGBM melakukan pemangkasan pada pohon untuk mencegah overfitting dan meningkatkan generalisasi pada data baru.
6. **Penggabungan Pohon**:  LightGBM menggabungkan hasil dari semua pohon yang telah dibangun. Pada tahap ini, setiap prediksi pohon diberi bobot berdasarkan performa pohon tersebut.

Model LightGBM menunjukkan performa lebih baik dalam klasifikasi dataset, ditandai dengan skor Recall yang lebih tinggi setelah tuning parameter. Konsistensi performanya antara train set dan test set juga teramati.

```python
# Pastikan preprocessor sudah terdefinisi
# Kolom kategorikal dan numerikal
categorical_cols = ['job', 'housing', 'loan', 'contact', 'month', 'poutcome']
numerical_cols = ['age', 'balance', 'pdays', 'campaign']

# Membuat transformasi kolom numerikal dan kategorikal
preprocessor = ColumnTransformer(
    transformers=[
        ('num', RobustScaler(), numerical_cols),  # Scaling numerikal
        ('cat', OneHotEncoder(sparse_output=False), categorical_cols)   # One-hot encoding untuk kategorikal
    ],
    remainder='drop'  # Hanya gunakan kolom yang terdaftar
)

# Pastikan preprocessor sudah dilatih pada data train
preprocessor.fit(X_train)

# Transformasi data test
X_test_transformed = preprocessor.transform(X_test)

# Ambil nama fitur numerikal
numerical_features = numerical_cols

# Ambil nama fitur kategorikal yang dihasilkan oleh OneHotEncoder
categorical_transformer = preprocessor.named_transformers_['cat']
categorical_features = categorical_transformer.get_feature_names_out(categorical_cols)

# Gabungkan nama fitur numerikal dan kategorikal untuk membentuk nama kolom final
feature_names = np.concatenate([numerical_features, categorical_features])

# Konversi hasil transformasi menjadi DataFrame dengan nama kolom
X_test_transformed = pd.DataFrame(X_test_transformed, columns=feature_names, index=X_test.index)

# Hitung permutation feature importance
perm_importance = permutation_importance(
    lgbm_tuned.named_steps['model'],  # Model tanpa preprocessing
    X_test_transformed,            # Data setelah preprocessing
    y_test,                        # Target asli
    scoring='recall',
    n_repeats=30,
    random_state=1515
)

# Buat DataFrame untuk hasil
perm_importance_df = pd.DataFrame({
    'Feature': feature_names,  # Nama fitur setelah preprocessing
    'Importance': perm_importance['importances_mean'],
    'Std': perm_importance['importances_std']
}).sort_values(by='Importance', ascending=False)

# Visualisasi feature importance

plt.figure(figsize=(10, 6))
plt.barh(perm_importance_df['Feature'], perm_importance_df['Importance'], color='lightcoral')
plt.errorbar(
    perm_importance_df['Importance'], perm_importance_df['Feature'],
    xerr=perm_importance_df['Std'], fmt='o', color='black', ecolor='gray', alpha=0.8
)
plt.gca().invert_yaxis()  # Membalikkan urutan supaya yang terbesar di atas
plt.title('Permutation Feature Importance (PFI)')
plt.xlabel('Importance Score')
plt.ylabel('Features')
plt.show()

```

#### Confusion Matrix

```python
categorical_cols = ['job', 'housing', 'loan', 'contact', 'month', 'poutcome']
numerical_cols = ['age', 'balance', 'pdays', 'campaign']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', RobustScaler(), numerical_cols),  # Scaling numerikal
        ('cat', OneHotEncoder(), categorical_cols)   # One-hot encoding untuk kategorikal
    ])

# Buat pipeline dengan model LGBMClassifier
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LGBMClassifier())
])

# Latih model tanpa tuning
model.fit(X_train, y_train)

# Prediksi pada data test sebelum tuning
y_pred_before = model.predict(X_test)

# Hitung confusion matrix sebelum tuning
cm_before = confusion_matrix(y_test, y_pred_before)

# Tuning model menggunakan GridSearchCV (contoh tuning untuk n_estimators dan learning_rate)
param_grid = {
    'classifier__n_estimators': [50, 100, 200],
    'classifier__learning_rate': [0.01, 0.1, 0.5]
}

# GridSearch untuk tuning
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='recall', verbose=1)

# Latih model setelah tuning
grid_search.fit(X_train, y_train)

# Prediksi pada data test setelah tuning
y_pred_after = grid_search.predict(X_test)

# Hitung confusion matrix setelah tuning
cm_after = confusion_matrix(y_test, y_pred_after)

# Visualisasikan confusion matrix sebelum dan setelah tuning berdampingan
fig, axes = plt.subplots(1, 2, figsize=(16, 6))  # Membuat subplots berdampingan

# Plot confusion matrix sebelum tuning
sns.heatmap(cm_before, annot=True, fmt='g', cmap='Blues', xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'], ax=axes[0])
axes[0].set_title('Confusion Matrix Sebelum Tuning')
axes[0].set_xlabel('Predicted Labels')
axes[0].set_ylabel('True Labels')

# Plot confusion matrix setelah tuning
sns.heatmap(cm_after, annot=True, fmt='g', cmap='Blues', xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'], ax=axes[1])
axes[1].set_title('Confusion Matrix Setelah Tuning')
axes[1].set_xlabel('Predicted Labels')
axes[1].set_ylabel('True Labels')

# Tampilkan hasil
plt.tight_layout()  # Agar plot tidak saling tumpang tindih
plt.show()
```
